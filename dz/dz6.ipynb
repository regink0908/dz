```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Titanic-Dataset.csv")
df.head()
```

```python
df_new = df.replace({'Sex':{'male':1, 'female':0}})
df_cor = df_new.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sbn

corrP = df_cor.corr(method='pearson', numeric_only=True).round(2)
sbn.heatmap(corrP, annot=True, vmax=1, vmin=-1, center=0)
plt.show()
```

```python
gf1 = df_cor.groupby(['Pclass', 'Survived'])['Survived'].count().reset_index(name='count')
sbn.barplot(x='Pclass', y='count', hue='Survived', data=gf1)
```

```python
gf1 = df_cor.groupby(['Sex', 'Survived'])['Survived'].count().reset_index(name='count')
sbn.barplot(x='Sex', y='count', hue='Survived', data=gf1)
```

```python
gf1 = df_cor.groupby(['SibSp', 'Survived'])['Survived'].count().reset_index(name='count')
sbn.barplot(x='SibSp', y='count', hue='Survived', data=gf1)
```

```python
gf1 = df_cor.groupby(['Parch', 'Survived'])['Survived'].count().reset_index(name='count')
sbn.barplot(x='Parch', y='count', hue='Survived', data=gf1)
```

```python
import numpy as np
gf1 = df_cor.groupby([pd.cut(df['Fare'], np.arange(0, 600, 100)), 'Survived'])['Survived'].count().reset_index(name='count')
sbn.barplot(x='Fare', y='count', hue='Survived', data=gf1)
```

```python
gf1 = df_cor.groupby([pd.cut(df['Age'], np.arange(0, 101, 10)), 'Survived'])['Survived'].count().reset_index(name='count')
sbn.barplot(x='Age', y='count', hue='Survived', data=gf1)
```
